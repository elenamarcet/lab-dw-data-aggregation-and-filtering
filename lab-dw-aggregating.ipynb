# Lab | Data Aggregation and Filtering

In this challenge, we will continue to work with customer data from an insurance company. We will use the dataset called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"

df = pd.read_csv(url)
```

```python
df
```

```python
#paso las columnas a minuscula y reemplazo los espacios por _

df.columns = df.columns.str.lower().str.replace(' ', '_')

df.columns
```

```python
#Nuevo Data Frame donde este el total claim amount mayor a 1000 y la response sea Yes

new_df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

new_df
```

```python

#Guardo el nuevo data frame en un archivo csv nuevo 
new_df.to_csv("clientes_filtrados.csv", index=False)
```

```python
#Filtrar para obtener solo los clientes que responden Yes

yes_response = df[df['response'] == 'Yes']

#Calcular el promedio de total claim amount por cada policy type y gender
average_claim_amount = yes_response.groupby(['policy_type', 'gender'])['total_claim_amount'].mean().reset_index()

average_claim_amount
```

```python
#3

#Contar el numero de clientes por estado
customers_state = df['state'].value_counts().reset_index()
customers_state.columns = ['state', 'customer_count']

#Incluir solo los estados que tengan mas de 500 clientes
states_ = customers_state[customers_state['customer_count'] > 500]

states_
```

```python
#4:

#Convertir customer_lifetime_value en valores numericos
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

#Agrupar y calcular estadisticas por nivel de educacion y genero
stadistics_ = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).reset_index()

#Renombrar las columnas para que sea mas claro
stadistics_.columns = ['education', 'gender', 'max_customer_lifetime_value', 'min_customer_lifetime_value', 'median_customer_lifetime_value' ]

stadistics_
```
